```python
#Librerias
import numpy as np
from numpy import linalg as la
import math
import sympy
import random
import matplotlib.pyplot as plt
from numpy import pi,sin,cos,e
from matplotlib import pyplot as plt

#Paso 1
L=800
N=2000
K=1201

A=np.linspace(-3*pi,3*pi, num=N)
aux=list()
aux1=list()
for j in range(0,N):
    ch=random.uniform(0,1/10)
    ft=sin(A[j])+(ch*sin(50*(A[j])))
    ft1=sin(A[j])
    aux.append(ft)
    aux1.append(ft1)
    
#paso 2
X=np.zeros((L,K))
for k in range(0,L):
    X[k,:]=aux[k:k+K]

#paso 3
u,s1,vh=np.linalg.svd(X)

s2=s1.copy()

#paso 4
t1=(np.max(s1)-np.min(s1))/L

for i in range(0,L-2):
    e=s1[i]-s1[i+1]
    e1=s1[i+1]-s1[i+2]

    if e<t1 and e1<t1:
        r=i-2
        i=i
        print("El menor valor de i es i=" + str(i))
        print("Por lo tanto el r buscado es r=" + str(r))
        break

for k in range(0,L):
    if k>=r:
        s2[k]=0
   
s=np.zeros((L,K))
s[0:L,0:L]=np.diag(s2)

print("\n") 
print("Observe las dimenciones de la matriz X son: "+ str(np.shape(X)))
print("Observe las dimenciones de la matriz U son: "+ str(np.shape(u)))
print("Observe las dimenciones de la matriz Sigma son: "+ str(np.shape(s)))
print("Observe las dimenciones de la matriz V son: "+ str(np.shape(vh)))

def Mul(A,B):
    D=np.dot(A,B)
    return(D)
A1=Mul(u,s)
Xr=Mul(A1,vh)
print("Observe las dimenciones de la matriz X_r son: "+ str(np.shape(Xr)))

#paso 5
print("\n") 
yr=list()
for i in range(0,N):
    
    if 0<=i<L-1:
        ayuda=list()
        for j in range(0,i):
            out=Xr[j,i-j]
            ayuda.append(out)
        yr.append(sum(ayuda)/(i+1))

    elif L-1<=i<K-1:
        ayuda1=list()
        for j in range(0,L):
            out=Xr[j,i-j]
            ayuda1.append(out)
        yr.append(sum(ayuda1)/L)
        
    elif K-1<i:
        ayuda2=list()
        for j in range(i-K+1,N-K+1):
            out=Xr[j,i-j]
            ayuda2.append(out)
        yr.append(sum(ayuda2)/N-i)

#paso 6
plt.figure(figsize=(9,6))
plt.plot(aux1)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Función Original")
plt.show()

print("\n")
plt.figure(figsize=(9,6))
plt.plot(aux)
plt.ylim(-1.5,1.5)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Función Ruidosa")
plt.show()

print("\n")
plt.figure(figsize=(10,6))
plt.plot(yr)
plt.xlim(0,1199)
plt.ylim(-1.5,1.5)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de la Función Recuparada")
plt.show()

print("\n")
plt.figure(figsize=(10,6))
plt.plot(yr)
plt.plot(aux)
plt.xlim(0,1200)
plt.ylim(-1.5,1.5)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de la Función Recuparada sobre la Función Ruidosa")
plt.show()

#paso 7
print("\n")
plt.figure(figsize=(7,6))
plt.plot(s1)
plt.xlim(0,10)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de todos los Valores Propios")
plt.show()

print("\n")
plt.figure(figsize=(7,6))
plt.plot(s2)
plt.xlim(0,10)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de los Valores Propios Restantes")
plt.show()

print("\n")
plt.figure(figsize=(7,6))
plt.plot(s2)
plt.plot(s1)
plt.xlim(0,10)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de los Valores Propios Sobre los Valores Propios Restantes")
plt.show()
print("\n")
plt.figure(figsize=(10,6))
plt.plot(yr)
plt.plot(aux)
plt.xlim(0,800)
plt.ylim(-1.5,1.5)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de la Función Recuparada sobre la Función Ruidosa")
plt.show()

print("\n")
plt.figure(figsize=(7,6))
plt.plot(s1)
plt.xlim(0,10)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de todos los Valores Propios")

print("\n")
plt.figure(figsize=(7,6))
plt.plot(s2)
plt.xlim(0,10)
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
plt.title("Gráfica de los Valores Propios Restantes")
plt.show()
```
